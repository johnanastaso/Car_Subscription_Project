```python
#calculation of the final price for Avis cars
def calculate_final_price(row):
    if row['leasing_company'] != 'Avis':
        return row['price']
    else:
        price = row['price']
        if price < 300:
            return price + price * 0.2
        elif 300 <= price < 450:
            return price + price * 0.248
        elif 450 <= price < 600:
            return price + price * 0.252
        elif 600 <= price < 900:
            return price + price * 0.262
        else:
            return price + price * 0.303
```

```python
#apply calculation on greek market dataframe
greek_market_df['final_price'] = greek_market_df.apply(calculate_final_price, axis=1)
```

```python
#apply calculation on total market dataframe
total_market_df['final_price'] = total_market_df.apply(calculate_final_price, axis=1)
```
