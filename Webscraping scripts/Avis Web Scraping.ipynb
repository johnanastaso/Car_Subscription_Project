Function that creates a dictionary for each car with its characteristics

```python
from datetime import datetime
import numpy as np
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from bs4 import BeautifulSoup

def scrape_page_selenium(link):
    driver_path = r"C:\Users\Giannis Anastaso\Desktop\Useful tools\geckodriver.exe"
    
    service = Service(driver_path)
    
    driver = webdriver.Firefox(service=service)
    
    try:
        driver.get(link)
        driver.implicitly_wait(3)
        
        html = driver.page_source
        
        soup = BeautifulSoup(html, 'html.parser')
        
        try:
            brand = soup.find('div', {'class': 'col text-center'}).find('h1').text.strip().split()[0]
        except:
            brand = np.nan
        try:
            model = soup.find('div', {'class': 'col text-center'}).find('h1').text.strip().split()[1]
        except:
            model = np.nan
        try:
            price = soup.find('p', {'class': 'price mb-2'}).text.strip().split('€')[0]
        except:
            price = np.nan
            print('Not found')
        try:
            displacement = soup.find_all('div', {'class': 'detail'})[1].find('p').text
        except:
            displacement = np.nan
        try:
            horsepower = soup.find_all('div', {'class': 'detail'})[2].find('p').text.split('/')[1]
        except:
            horsepower = np.nan
        try:
            gearbox = soup.find_all('div', {'class': 'detail'})[4].find('p').text
        except:
            gearbox = np.nan
        try:
            fuel = soup.find_all('div', {'class': 'detail'})[3].find('p').text
        except:
            fuel = np.nan
        
        # Create a dictionary with scraped data
        scrape_data = {
            'date': datetime.now(),
            'link': link,
            'brand': brand,
            'model': model,
            'price': price,
            'displacement': displacement,
            'horsepower': horsepower,
            'gearbox': gearbox,
            'fuel': fuel
        }

        return scrape_data

    except Exception as e:
        print(f'An error occurred: {e}')

    finally:
        
        driver.quit()

# Example usage:
car_url = car_url_list[0]
result = scrape_page_selenium(car_url)
print(result)
```

```python

from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from bs4 import BeautifulSoup

def scrape_page_selenium(link):
    driver_path = r"C:\Users\Giannis Anastaso\Desktop\Useful tools\geckodriver.exe"
    
    # Initialize the service with the executable path
    service = Service(driver_path)
    
    # Initialize the driver with the service
    driver = webdriver.Firefox(service=service)
    
    try:
        driver.get(link)
        
        # Wait for 5 seconds to let the page load
        time.sleep(5)
        
        # Use Selenium to get the page source
        html = driver.page_source
        
        # Parse the HTML with BeautifulSoup
        soup = BeautifulSoup(html, 'html.parser')
        
        try:
            brand = soup.find('div', {'class': 'col text-center'}).find('h1').text.strip().split()[0]
        except:
            brand = np.nan
        try:
            model = soup.find('div', {'class': 'col text-center'}).find('h1').text.strip().split()[1]
        except:
            model = np.nan
        try:
            price = soup.find('p', {'class': 'price mb-2'}).text.strip().split('€')[0]
        except:
            price = np.nan
            print('Not found')
        try:
            displacement = soup.find_all('div', {'class': 'detail'})[1].find('p').text
        except:
            displacement = np.nan
        try:
            horsepower = soup.find_all('div', {'class': 'detail'})[2].find('p').text.split('/')[1]
        except:
            horsepower = np.nan
        try:
            gearbox = soup.find_all('div', {'class': 'detail'})[4].find('p').text
        except:
            gearbox = np.nan
        try:
            fuel = soup.find_all('div', {'class': 'detail'})[3].find('p').text
        except:
            fuel = np.nan
        
        # Create a dictionary with scraped data
        scrape_data = {
            'date': datetime.now(),
            'link': link,
            'brand': brand,
            'model': model,
            'price': price,
            'displacement': displacement,
            'horsepower': horsepower,
            'gearbox': gearbox,
            'fuel': fuel
        }

        return scrape_data

    except Exception as e:
        print(f'An error occurred: {e}')

    finally:
        # Close the WebDriver
        driver.quit()
```

```python
scraped_data_list = []

# Loop through each car URL
for car_url in car_url_list[:20]:
    result = scrape_page_selenium(car_url)
    scraped_data_list.append(result)

# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(scraped_data_list)
df
```

```python
# Filter out None values or dictionaries with missing keys
filtered_data_list = [item for item in scraped_data_list_2 if item is not None and item]

# If the list is not empty, get the union of all keys in the dictionaries
if filtered_data_list:
    all_keys = list(set().union(*[d.keys() for d in filtered_data_list]))
    
    # Convert the filtered list to a DataFrame
    df2 = pd.DataFrame(filtered_data_list, columns=all_keys)
else:
    # Handle the case where the list is empty
    df2 = pd.DataFrame()


df2
```

```python
scraped_data_list_3 = []
failed_urls = []

# Loop through each car URL for the final 1603 cars
for car_url in car_url_list[1500:3103]: 
    result = scrape_page_selenium(car_url)
    
    if result is not None and result:  # Check if the result is not None and not an empty dictionary
        scraped_data_list_3.append(result)
    else:
        failed_urls.append(car_url)
        print(f"Failed to scrape data for URL: {car_url}")

df3 = pd.DataFrame(scraped_data_list_3)

df3

# Display the list of failed URLs
print("Failed URLs:", failed_urls)
```
