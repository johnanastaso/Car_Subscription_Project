```python
def extract_brand_and_model(input_string):
    # Define a regular expression pattern to match the brand and model
    pattern = re.compile(r'(?P<brand>[A-Za-z]+)\s(?P<model>.+)')
    
    # Use the pattern to search for matches in the input string
    match = pattern.match(input_string)
    
    # Check if a match is found
    if match:
        # Extract the brand and model from the named groups in the pattern
        brand = match.group('brand')
        model = match.group('model')
        
        return brand, model
    else:
        # Return None if no match is found
        return None
```

```python
def convert_str_to_int(x):
    pattern = r'\b(\d+)\b'
    match = re.search(pattern, x)
    if match:
        return int(match.group(1))
    else:
        print("No match found.")
        return None
```

### The website has 7 pages with hrefs, so we parse them in a list to iterate through all of the cars

```python
pages_urls = ['https://flexcar.gr/cars/']
page = '?page='
i = 2
for item in range (1, 7):
    selida = url + page + str(i)
    pages_urls.append(selida)
    i += 1
pages_urls
```

```python
inside_car_url = []
pattern = re.compile(r'/cars/\w+/\w+')
for page in pages_urls:
    response_ = requests.get(page)
    if response_.status_code == 200:
        car_soup = BeautifulSoup(response_.content, "html.parser")
        hrefs_ = [a['href'] for a in car_soup.find_all('a', href=True) if '/cars/' in a['href']]
        filtered_hrefs = [href for href in hrefs_ if pattern.search(href)]
        for href in filtered_hrefs:
            inside_car_url.append(main_link + href)
    else:
        print('error')
```

```python
def extract_brand_and_model(input_string):
    # Define a regular expression pattern to match the brand and model
    pattern = re.compile(r'(?P<brand>[A-Za-z]+)\s(?P<model>.+)')
    
    # Use the pattern to search for matches in the input string
    match = pattern.match(input_string)
    
    # Check if a match is found
    if match:
        # Extract the brand and model from the named groups in the pattern
        brand = match.group('brand')
        model = match.group('model').strip()
        
        return brand, model
    else:
        # Return None if no match is found
        return None, None

def cars_func(link):
    car_data = {}
    
    try:
        response_car = requests.get(link)
        soup_car = BeautifulSoup(response_car.text, 'html.parser')
        
        head = soup_car.find('h2', class_='f700 dark h5')
        car_name = head.text.strip().replace('\n', "")
        
        # Use extract_brand_and_model function to get brand and model
        brand, model = extract_brand_and_model(car_name)
        
        # Assign brand and model to car_data dictionary
        car_data['link'] = link
        car_data['brand'] = brand
        car_data['model'] = model
        car_info = soup_car.find_all('span', class_='dark f500 pxs')
        info = [info.text for info in car_info]
        horsepower = info[0]
        car_data['horsepower'] = horsepower
        cc = info[1]
        car_data['cc'] = cc
        car_data['consumption'] = car_info[2].text if len(car_info) > 2 else np.nan
        fuel_transmission = car_info[3].text.split()
        car_data['fuel'] = fuel_transmission[0] if fuel_transmission else np.nan
        car_data['transmission'] = fuel_transmission[2] if len(fuel_transmission) > 2 else np.nan
        price = soup_car.find('div', class_='d-flex flex-column justify-content-start align-items-start').text.split()[0] if soup_car.find('div', class_='d-flex flex-column justify-content-start align-items-start') else np.nan
        price = price.replace(".", "")
        price = int(price)
        car_data['price'] = price

    except Exception as e:
        print('Error:', e)
    
    return car_data
```

### Testing if the function works and returns a dictionary with the car info for every unique link

```python
cars_func (inside_car_url[29])
```

```python
df = pd.DataFrame()
for url in inside_car_url:
    car =  cars_func(url)
    car_df = pd.DataFrame([car])
    df = pd.concat([df , car_df] ,  axis = 0 , ignore_index = True)
```
